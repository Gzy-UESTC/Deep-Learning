## 数据操作

```python
import torch

x = torch.arange(12, dtype=torch.float32).reshape((3, 4))
y = torch.tensor([[2.0, 1, 4, 3],
                  [1, 2, 3, 4],
                  [4, 3, 2, 1]])

print(torch.cat((x, y), dim=0)) # 上下链接
print(torch.cat((x, y), dim=1)) # 左右链接

# 广播机制
a = torch.arange(3).reshape(3,1)
b = torch.arange(2).reshape(1,2)
print(a, b, a+b)

# 访问
print(x[-1], x[1:3], x[1:3,2:3])

# numpy → tensor
A = x.numpy()
B = torch.tensor(A)
a = torch.tensor([3.5])
a = a.item()
print(type(A), type(B), a)

```
