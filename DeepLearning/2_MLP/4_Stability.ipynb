# Stability
## 梯度消失与爆炸
### 梯度消失
梯度从输出层出发，穿过一层层权重矩阵到达输入层

梯度消失 (Vanishing Gradients)

现象：靠近输入层的参数更新极慢，模型无法学习

元凶：Sigmoid激活函数

原理：Sigmoid在输入过大或过小时，导数接近于 0，在多层相乘后，梯度会以指数级速度萎缩

### 梯度爆炸
梯度爆炸 (Exploding Gradients)

现象：权重更新变得极大（甚至变成 NaN），模型直接崩溃

元凶：权重初始化值过大或网络层数过深

原理：如果每个层的权重矩阵特征值大于 1，100层累乘后的数值将是天文数字

## 打破对称性
如果所有权重初始值都设为 0（或任意相同常数）会怎样？

前向传播：同一层的所有隐藏单元接收相同的输入，计算出完全相同的输出

反向传播：它们得到的梯度也完全一样

结果：无论训练多久，这些神经元就像“多胞胎”一样同步动作，神经网络退化成了只有一个神经元的线性模型

解决方案：通过随机初始化赋予每个神经元不同的初速度，让它们朝不同的方向学习

## Xavier 初始化
为了平衡消失和爆炸，我们需要一种方案，让信息在经过每一层时，其方差（Variance）保持稳定

Xavier 初始化的目标是：让每一层输出的方差等于输入的方差

为了达到这个平衡，权重的方差应该取决于该层的输入维度 $n_{in}$ 和输出维度 $n_{out}$，具体的数学准则是：
$$\text{Var}(w) = \frac{2}{n_{in} + n_{out}}$$

如果是正态分布：抽样分布为 $N(0, \sqrt{\frac{2}{n_{in} + n_{out}}})$

如果是均匀分布：抽样区间为 $[-\sqrt{\frac{6}{n_{in} + n_{out}}}, \sqrt{\frac{6}{n_{in} + n_{out}}}]$
