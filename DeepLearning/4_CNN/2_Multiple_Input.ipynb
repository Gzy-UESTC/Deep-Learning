# Multiple Input and Output Channels

在之前的简单卷积中，我们通常假设输入是二维的（比如只有灰度的一张图像），但实际应用中，数据往往有更高的维度。例如，一张彩色的图片包含红（R）、绿（G）、蓝（B）三个颜色通道。这就引出了多输入多输出通道的概念

## 多输入通道
当你有一张彩色图片时，它的输入是一个三维张量，形状可以表示为 $c_i \times h \times w$（通道数 $\times$ 高度 $\times$ 宽度），这里的输入通道数 $c_i = 3$

为了处理多通道输入，我们的卷积核也必须有相同数量的输入通道。如果输入有 $c_i$ 个通道，那么卷积核的形状就会变成 $c_i \times k_h \times k_w$

多输入通道的卷积，实际上是各个通道分别做卷积后，再把结果合并（相加）在一起，提取的是所有通道融合后的特征



## 多输出通道

在神经网络中，随着层数的加深，图像的宽和高通常会变小，但提取的特征会越来越抽象。为了捕捉不同类型的特征，我们需要增加输出通道的数量 $c_o$

为了得到 $c_o$ 个输出通道，我们需要为每一个输出通道准备一个形状为 $c_i \times k_h \times k_w$ 的卷积核。
因此，整个卷积层的权重参数将变成一个四维张量，形状为：$c_o \times c_i \times k_h \times k_w$
