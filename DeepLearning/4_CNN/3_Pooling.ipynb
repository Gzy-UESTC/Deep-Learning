# Pooling Layer
## 为什么需要汇聚层
在图像处理中，我们会面临两个实际问题：

位置敏感性： 假设你要识别照片里的猫，这只猫向左移动了一个像素。虽然对人眼来说没区别，但对卷积层来说，由于像素位置变了，它输出的特征图的位置也会跟着移动。我们希望网络能有一种“平移不变性”，即不管猫在哪个确切位置，都能识别出来

计算量庞大： 随着网络变深，特征图的数量（通道数）越来越多。如果不把图像的宽和高缩小（降采样），后面的计算量会大到计算机受不了

## Maximum Pooling
汇聚层运算符由一个固定形状的窗口组成，该窗口根据其步幅大小在输入的所有区域上滑动，为固定形状窗口遍历的每个位置计算一个输出



## Average Pooling
每次滑动到一个位置，它计算窗口内所有值的平均数作为输出

通常用于保留背景信息，或者在网络的最末端用来把整个特征图压缩成一个值

## Padding and Stride
在深度学习中，汇聚层最常见的设置是窗口大小等于步幅大小

每经过一次这样的汇聚层，特征图的高度和宽度都会精准减半，面积变为原来的四分之一

```python
import torch
from torch import nn

# 假设输入是一个批次大小为1，3个通道，高宽为 4x4 的特征图
X = torch.arange(48, dtype=torch.float32).view((1, 3, 4, 4))

# 定义一个最大汇聚层：窗口大小为 2x2。
pool2d = nn.MaxPool2d(kernel_size=2)

# 前向传播
Y = pool2d(X)

print(X.shape) 
print(Y.shape) 
```
