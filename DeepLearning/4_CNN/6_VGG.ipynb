# VGG
为了提取更好的特征，我们应该用更大的卷积核，还是更深的网络？

VGG 的答案是：全部用最小的 $3 \times 3$ 卷积核，然后把网络做深

为什么连续的小卷积比一个大卷积更好?

感受野（Receptive Field）相同：2 个级联的 $3 \times 3$ 卷积层，其感受野等价于 1 个 $5 \times 5$ 卷积层；3 个级联的 $3 \times 3$ 卷积层，等价于 1 个 $7 \times 7$ 卷积层

非线性更强：3 个 $3 \times 3$ 卷积层意味着会经过 3 次 ReLU 激活函数，而 1 个 $7 \times 7$ 只有 1 次。更多的非线性让网络能学习到更复杂的特征

参数量更少

## VGG 模块化设计
一个基础的VGG块包含：

$N$ 个连续的卷积层：全部固定使用 $3 \times 3$ 卷积核，填充（padding）为 $1$（保持分辨率不变）

ReLU 激活函数：跟在每个卷积层后面

$1$ 个最大池化层：固定使用 $2 \times 2$ 窗口，步幅为 $2$
