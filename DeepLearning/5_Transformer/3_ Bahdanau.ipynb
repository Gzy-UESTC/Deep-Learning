# Bahdana
在 Bahdanau 提出注意力机制之前，经典的机器翻译模型是纯粹的 RNN Encoder-Decoder 架构：

Encoder：把一整句极长的英文（比如 50 个词），疯狂压缩、最后融合成唯一的一个、固定长度的上下文向量 $c$

Decoder：拿着这个唯一的向量 $c$，从头到尾去生成中文

句子越长，前面词的信息丢失就越严重

## Bahdanau：回头看机制
让 Decoder 在每次翻译一个新词的时候，都动态地回头去翻阅原句

Query (查询 $Q$)：Decoder 当前时刻的前一个隐藏状态 $s_{t-1}$

Key (键 $K$)：Encoder 所有时间步的隐藏状态 $h_1, h_2, \dots, h_T$

Value (值 $V$)：与 Key 完全相同，也是 Encoder 所有时间步的隐藏状态 $h_1, h_2, \dots, h_T$

加性注意力太慢了： 包含了太多的权重矩阵乘法，远不如缩放点积来得简单暴力

RNN 无法并行： 无论怎么加注意力，Decoder 还是得一个词一个词、串行地生成状态 $s_{t-1}$。无法并行，就意味着无法彻底榨干 GPU 的几千个流处理器
