# Self-Attention & Positional Encoding
## Self-Attention
在之前的多头注意力中，我们提到了查询（Query）、键（Key）和值（Value）

如果 Query 来自一个句子，Key 和 Value 来自另一个句子，这叫交叉注意力

如果 Query、Key 和 Value 全部来自同一个输入序列本身，这就叫自注意力

### 核心逻辑
假设我们有一个输入序列 $X = [x_1, x_2, \dots, x_n]$，在自注意力中：
$$Q = XW^{(q)}$$
$$K = XW^{(k)}$$
$$V = XW^{(v)}$$

### 为什么需要自注意力
RNN 的痛点：必须按顺序一个词一个词地计算（无法并行），且相隔极远的两个词，信息传递会逐渐衰减

CNN 的痛点：每次卷积只能看到局部的一个小窗口，想要看到整个句子的关联，得堆叠很多层

自注意力：它的感受野是全局的。无论两个词在句子中隔得有多远，它们在自注意力中都只需要一步计算就能建立直接联系。而且，整个句子可以通过矩阵乘法高度并行地一次性算完

## 位置编码
自注意力的核心计算是 $Q \times K^\top$。在这个矩阵乘法里，无论你是把一句话正着放还是反着放，词与词之间的点积结果（相关度）是完全一样的

为了让模型知道谁在前面，谁在后面，Transformer 采用了最简单的办法：在输入模型之前，直接把位置信息加到词向量里

假设输入序列包含 $n$ 个词，第 $i$ 个词（$i=0, 1, \dots, n-1$）的原始词嵌入向量为 $x_i \in \mathbb{R}^d$

为每个位置 $i$ 凭空捏造出一个位置向量 $p_i \in \mathbb{R}^d$。最终送入自注意力的实际输入是：$$X_{input} = X + P$$

对于位置向量 $p_i$ 中的第 $j$ 个维度:

偶数维度（$2j$） 使用正弦函数：
$$p_{i, 2j} = \sin\left(\frac{i}{10000^{2j/d}}\right)$$
奇数维度（$2j+1$） 使用余弦函数：
$$p_{i, 2j+1} = \cos\left(\frac{i}{10000^{2j/d}}\right)$$
