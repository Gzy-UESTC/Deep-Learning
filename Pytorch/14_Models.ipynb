# VGG 16

```python
import torchvision
import torch.nn as nn
import torch

train_data = torchvision.datasets.CIFAR10('./data', train=True, download=True) # Download CIFAR10 training dataset

vgg16_false = torchvision.models.vgg16(pretrained=False) # Load VGG16 model without pre-trained weights
vgg16_true = torchvision.models.vgg16(pretrained=True) # Load VGG16 model with pre-trained weights

# VGG16 分1000类，而 CIFAR10 只有10类，因此需要修改最后的全连接层
vgg16_true.classifier.add_module('add_linear', nn.Linear(1000, 10)) # 添加一个新的线性层，将输出类别数从1000改为10
vgg16_true.classifier[6] = nn.Linear(4096, 10) # 直接替换最后一层全连接层

# 保存/加载模型
# 方式一：保存整个模型
torch.save(vgg16_true, 'vgg16_cifar10_full.pth')
torch.load('vgg16_cifar10_full.pth')

# 方式二：只保存模型参数
torch.save(vgg16_true.state_dict(), 'vgg16_cifar10_params.pth')
vgg16 = torchvision.models.vgg16(pretrained=False)
vgg16.load_state_dict(torch.load('vgg16_cifar10_params.pth')) # 加载参数到新模型
```
