## Dataset

获取每一个数据及其 label

获取总共有多少数据

```python
from torch.utils.data import Dataset
import os
from PIL import Image

class MyData(Dataset):
    def __init__(self, root_dir, label_dir):
        self.root_dir = root_dir
        self.label_dir = label_dir
        self.path = os.path.join(root_dir, label_dir) # 将根目录和标签目录拼接成完整的路径
        self.image_path = os.listdir(self.path) # 列出路径下的所有文件
    
    def __getitem__(self, idx): # 获取指定索引的数据项
        img_name = self.image_path[idx] # 获取指定索引的文件名
        img_item_path = os.path.join(self.root_dir, self.label_dir, img_name) # 拼接成完整的文件路径
        img = Image.open(img_item_path) # 打开图像文件
        label = self.label_dir # 标签就是目录名
        return img, label # 返回图像和标签

    def __len__(self): # 返回数据集的大小

        return len(self.image_path)

root_dir = "/home/m0nesy/Learning/Deep-Learning/Pytorch/data"
ant_label_dir = "ants"
bee_label_dir = "bees"
ants_dataset = MyData(root_dir, ant_label_dir) # 创建数据集实例
bees_dataset = MyData(root_dir, bee_label_dir) # 创建数据集实例

train_dataset = ants_dataset + bees_dataset # 合并两个数据集
```
