# Transforms

## ToTensor

```python
from torchvision import transforms
from PIL import Image
from torch.utils.tensorboard import SummaryWriter

img_path = "test.jpg"
img = Image.open(img_path)

tensor_transform = transforms.ToTensor() 
tensor_img = tensor_transform(img) # 转换为张量

writer = SummaryWriter("logs")
writer.add_image("test_img", tensor_img) # 将图像写入TensorBoard
writer.close()
```

## Normalize

```python
trans_norm = transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
img_norm = trans_norm(tensor_img) # 归一化处理
```

## Resize

```python
trans_resize = transforms.Resize((128, 128))
img_resize = trans_resize(img) # 调整图像大小(PIL图像)
img_resize_tensor = tensor_transform(img_resize) # 转换为张量

trans_resize_2 = transforms.Resize(128) # 调整图像短边为128像素
trans_compose = transforms.Compose([trans_resize_2, tensor_transform]) # 组合多个变换
```

## RandomCrop

```python
trans_random = transforms.RandomCrop(128) # 随机裁剪图像
trans_compose_2 = transforms.Compose([trans_random, tensor_transform]) # 组合随机裁剪和转换为张量
img_random = trans_compose_2(img) # 应用组合变换到图像
```
