# Dataset

```python
import torchvision
from torchvision import transforms
from torch.utils.tensorboard import SummaryWriter

pil_to_tensor = transforms.ToTensor() # 将PIL图像转换为张量
dataset_transform = transforms.Compose([pil_to_tensor]) # 定义数据集转换
train_set = torchvision.datasets.CIFAR10(root='./data', train=True, download=False) # 下载训练集
test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=False) # 下载测试集

writer = SummaryWriter('logs') # 创建SummaryWriter对象
for i in range(10): # 遍历前10张图像
    img, label = test_set[i] # 获取图像和标签
    img_tensor = dataset_transform(img) # 将图像转换为张量
    writer.add_image('test_set', img_tensor, i) # 将图像添加到TensorBoard中

writer.close() # 关闭SummaryWriter对象
```
