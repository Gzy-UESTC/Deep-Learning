# Dataloader

```python
import torchvision
from torch.utils.data import DataLoader

transform = torchvision.transforms.ToTensor()
test_set = torchvision.datasets.CIFAR10(root="./data", train=False, transform=transform, download=True)
test_loader = DataLoader(test_set, batch_size=4, shuffle=False, num_workers=2, drop_last=False) # batch_size 说明了每个批次中包含多少个样本，
                                                                                                # shuffle=False 表示不打乱数据，
                                                                                                # num_workers=2 表示使用两个子进程来加载数据，
                                                                                                # drop_last=False 表示如果最后一个批次的样本数量不足 batch_size，则不丢弃这些样本。
for data in test_loader:
    images, labels = data # 从数据加载器中获取一个批次的数据，images 是图像数据，labels 是对应的标签
    print(images.shape)  # 输出图像的形状
    print(labels)       # 输出标签
    break  # 只查看第一个批次的数据
```
