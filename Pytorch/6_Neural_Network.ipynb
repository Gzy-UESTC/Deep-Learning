# Neural Network

```python
from torch import nn
import torch

class Mynet(nn.Module):
    def __init__(self):
        super().__init__() # 调用父类的构造函数，初始化神经网络模块

    def forward(self, input): # 定义前向传播函数，接受输入数据 input
        output = input + 1
        return output

mynet = Mynet()
x = torch.tensor([1, 2, 3])
y = mynet(x)
print(y)
```
