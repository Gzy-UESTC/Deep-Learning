# Activation Function

```python
import torch
import torch.nn as nn
from torch.nn import ReLU

input = torch.tensor([[1, -0.5],
                      [-1, 3]])
input = torch.reshape(input,(-1, 1, 2, 2))  # 调整输入形状为 (batch_size, channels, height, width)

class Activation(nn.Module):
    def __init__(self):
        super(Activation, self).__init__()
        self.relu = ReLU() 

    def forward(self, x):
        x = self.relu(x) # 应用ReLU激活函数
        return x

mynet = Activation()
output = mynet(input)
print("input:", input)
print("output:", output)
```
